```python
class table_manager(object):
    '''
    使用table_list管理程式中所有scope的symbol table
    table_list中只會有當前的scope以及父scope的table
    '''
    def __init__(self):
        self.table_list = []
        
    def add_table(self, current_scope_number):
        '''呼叫此函式當visitor進入新的scope'''
        table = symbol_table(current_scope_number)
        self.table_list.append(table)
        self.current_table = table
        
    def pop_table(self):
        '''呼叫此函式當visitor離開scope'''
        try:
            table = self.table_list.pop()
            if len(self.table_list) is not 0:
                self.current_table = self.table_list[-1]
            else: 
                self.current_table = None
                return table;
    
        except:
            print('table list is empty')
            return
    
    def add_symbol(self,symbol):
        self.current_table.add_or_update(symbol)
        
    def get_symbol_value(self,symbol_name:str):
        try:
            return self.current_table.get_symbol_value(symbol_name)
        except:
            print('table list is empty')
        

class symbol_table(object):
    '''
    此class的目的是為了儲存每個scope中的symbol name 以及value
    將資訊存在table中
    table中儲存identifier這個類別的物件
    '''
    def __init__(self,scope_number):
        self.table = []
        self.num = scope_number #分辨此table屬於哪個scope，從0開始編號
        
    def add_or_update(self, symbol):
        '''
        確認table中是否出現過該symbol
        如果有:替換該symbol 的value
        無:將symbol存入table
           '''
        if not len(self.table):
            self.table.append(symbol)

        for i in self.table: #有找到
            if i.name == symbol.name:
                i.value = symbol.value
                return
            
        self.table.append(symbol) #沒找到
        
    def get_symbol_value(self,symbol_name:str):
        for i in self.table:
            if i.name == symbol_name:
                return i.value
        else:
            return "not found"
       
class symbol():
    '''
    存放著symbol的名字以及被賦的值
    皆以字串保存    
    '''
    def __init__(self, name:str, value:str):
        self.name = name
        self.value = value
    
```

```python
manager = table_manager()
current_scope_number = 0
manager.add_table(current_scope_number)
current_scope_number+=1
a = symbol('n','n')
b = symbol('m','m')
manager.add_symbol(a)
manager.add_symbol(b)
```
